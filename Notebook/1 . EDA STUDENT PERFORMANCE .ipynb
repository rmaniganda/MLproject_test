```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your DataFrame
# df = pd.read_csv('your_file.csv')  # Replace with your actual data loading code

# Print column names to verify
print("Original column names:", df.columns)

# Strip any leading/trailing spaces from column names
df.columns = df.columns.str.strip()
print("Cleaned column names:", df.columns)

# Check if the 'race/ethnicity' column exists
if 'race/ethnicity' in df.columns:
    # Group data by 'race/ethnicity'
    Group_data2 = df.groupby('race/ethnicity')
    
    # Plotting
    f, ax = plt.subplots(1, 3, figsize=(20, 8))

    # Plot 1: Math score
    sns.barplot(x=Group_data2['math score'].mean().index,
                y=Group_data2['math score'].mean().values,
                palette='mako',
                ax=ax[0])
    ax[0].set_title('Math Score', color='#005ce6', size=20)
    for container in ax[0].containers:
        ax[0].bar_label(container, color='black', size=15)

    # Plot 2: Reading score
    sns.barplot(x=Group_data2['reading score'].mean().index,
                y=Group_data2['reading score'].mean().values,
                palette='flare',
                ax=ax[1])
    ax[1].set_title('Reading Score', color='#005ce6', size=20)
    for container in ax[1].containers:
        ax[1].bar_label(container, color='black', size=15)

    # Plot 3: Writing score
    sns.barplot(x=Group_data2['writing score'].mean().index,
                y=Group_data2['writing score'].mean().values,
                palette='coolwarm',
                ax=ax[2])
    ax[2].set_title('Writing Score', color='#005ce6', size=20)
    for container in ax[2].containers:
        ax[2].bar_label(container, color='black', size=15)

    plt.tight_layout()
    plt.show()

else:
    print("Column 'race/ethnicity' not found. Please verify the column name.")
```

#### Insights 
- Group E students have scored the highest marks. 
- Group A students have scored the lowest marks. 
- Students from a lower Socioeconomic status have a lower avg in all course subjects

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your DataFrame
# df = pd.read_csv('your_file.csv')  # Replace with your actual data loading code

# Print column names to verify
print("Columns in DataFrame:", df.columns)

# Strip any leading/trailing spaces from column names
df.columns = df.columns.str.strip()
print("Cleaned column names:", df.columns)

# Check if 'parental level of education' column exists
if 'parental level of education' in df.columns:
    plt.rcParams['figure.figsize'] = (15, 9)
    plt.style.use('fivethirtyeight')
    sns.countplot(x='parental level of education', data=df, palette='Blues')
    plt.title('Comparison of Parental Education', fontweight='bold', fontsize=20)
    plt.xlabel('Degree')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("Column 'parental level of education' not found. Please verify the column name.")
```

#### BIVARIATE ANALYSIS ( Is parental education has any impact on student's performance ? )

```python
import matplotlib.pyplot as plt

# Ensure column names are correct
df.columns = df.columns.str.strip()

# Check if 'parental level of education' exists and numeric columns are available
if 'parental level of education' in df.columns and all(col in df.columns for col in ['math score', 'reading score', 'writing score']):
    grouped_data = df.groupby('parental level of education').agg({
        'math score': 'mean',
        'reading score': 'mean',
        'writing score': 'mean'
    })
    
    # Plot the mean of each numeric column
    grouped_data.plot(kind='barh', figsize=(10, 10))
    plt.title('Average Scores by Parental Level of Education', fontsize=16, fontweight='bold')
    plt.xlabel('Average Score')
    plt.ylabel('Parental Level of Education')
    plt.legend(title='Scores', bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0.)
    plt.tight_layout()
    plt.show()
else:
    print("Required columns are missing. Please check the column names.")
```

#### Insights 
- The score of student whose parents possess master and bachelor level education are higher than others.
